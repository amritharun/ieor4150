```r
# CHANGE TICKER HERE
# Options: aapl,googl
ticker <- 'aapl'
```

```r
file <- paste('data/',ticker,'.csv', sep="")
df <- read.csv(file)
df$Date <- as.Date(df$Date , format = "%m/%d/%y")
df <- df[df$Date >= "2020-02-01" & df$Date <= "2020-05-31",]
```

```r
df$Close.Last <- as.numeric(gsub('\\$|,', '', df$Close.Last))
df$Open <- as.numeric(gsub('\\$|,', '', df$Open))
data <- log(df$Open / df$Close.Last)
```
